```python
import plotly 
# only for offline
import plotly.graph_objects as go
import pandas as pd
```

```python
path_testdata = 'testset_full.csv'
#movie_data =pd.read_csv(path, sep='\t', head = 0)
test_data = pd.read_csv(path_testdata)
test_data
```

```python
path_g1 = 'test_result_final_group1.csv'
#movie_data =pd.read_csv(path, sep='\t', head = 0)
g1_result_final = pd.read_csv(path_g1)
#remove Unamed column
g1_result_final = g1_result_final.loc[:, ~g1_result_final.columns.str.contains('^Unnamed')]
g1_result_final = g1_result_final.drop_duplicates(subset=None, keep='first', inplace=False)
g1_result_final = g1_result_final.drop(columns=['prediction','pred_meta_1','pred_meta_2'])
g1_result_final = g1_result_final.rename(columns={"pred_video": "video_pre_g1", "pred_meta_3": "meta_pre_g1"})

g1_result_final
```

```python
path_g2 = 'Group_2.csv'
#movie_data =pd.read_csv(path, sep='\t', head = 0)
g2_result = pd.read_csv(path_g2)
g2_result = g2_result.drop(columns=['imdb_id','prediction'])
g2_result = g2_result.rename(columns={"metadata_pred":"meta_pre_g2","trailer_pred":"video_pre_g2"})
g2_result_final = pd.merge(g1_result_final,g2_result,how='outer',on='url')
g2_result_final = g2_result_final.drop_duplicates(subset=None, keep='first', inplace=False)
g2_result_final
```

```python
path_g3 = 'Group3.csv'
#movie_data =pd.read_csv(path, sep='\t', head = 0)
g3_result_final = pd.read_csv(path_g3)
g3_result_final = g3_result_final.drop_duplicates(subset=None, keep='first', inplace=False)
g3_result_final = g3_result_final.drop(columns=["prediction","ground_truth","name","imdb_id"])
g3_result_final = g3_result_final.rename(columns={"prediction_metadata":"meta_pre_g3","prediction_video":"video_pre_g3"})
g3_result_final = pd.merge(g2_result_final,g3_result_final,how='outer',on='url')
g3_result_final = g3_result_final.drop_duplicates(subset=None, keep='first', inplace=False)
g3_result_final
```

```python
path_meta_g4 = 'Group_4_meta_final.csv'
#movie_data =pd.read_csv(path, sep='\t', head = 0)
result_meta_data_g4 = pd.read_csv(path_meta_g4)
result_meta_data_g4 = result_meta_data_g4.drop(columns=['openingBox'])
result_meta_data_g4 = result_meta_data_g4.rename(columns={'prediction':'meta_pre_g4'})


path_video_g4 = 'Group_4_video_predictions92.csv'
#movie_data =pd.read_csv(path, sep='\t', head = 0)
result_video_data_g4 = pd.read_csv(path_video_g4)
result_video_data_g4 = result_video_data_g4.rename(columns={"prediction":"video_pre_g4"})

g4_result = pd.merge(result_meta_data_g4,result_video_data_g4,how='outer',on='imdb_id')
g4_result = g4_result.drop(columns=['imdb_id'])
g4_result_final = pd.merge(g3_result_final,g4_result,how='outer', on='url')
g4_result_final = g4_result_final.drop_duplicates(subset=None, keep='first', inplace=False)
g4_result_final
```

```python
path_g5 = 'Group_5.csv'
g5_result = pd.read_csv(path_g5)
g5_result = g5_result.drop(columns=['imdb_id','prediction','prediction_hybird'])
g5_result = g5_result.rename(columns={'prediction_video':'video_pre_g5','prediction_metadata':'meta_pre_g5'})
g5_result_final = pd.merge(g4_result_final,g5_result,how='outer', on='url')
g5_result_final = g5_result_final.drop_duplicates(subset=None, keep='first', inplace=False)
g5_result_final
```

```python
path_g6 = 'group6_predictions.csv'
g6_result = pd.read_csv(path_g6)
g6_result = g6_result.loc[:, ~g6_result.columns.str.contains('^Unnamed')]
g6_result = g6_result.drop(columns=['imdb_id'])
g6_result = g6_result.rename(columns={'metadata_prediction':'meta_pre_g6','trailer_prediction':'video_pre_g6'})
g6_result_final = pd.merge(g5_result_final,g6_result,how='outer', on='url')
g6_result_final = g6_result_final.drop_duplicates(subset=None, keep='first', inplace=False)
g6_result_final
```

```python
path_g7_meta = 'Group7_Meta.csv'
#movie_data =pd.read_csv(path, sep='\t', head = 0)
g7_result_meta = pd.read_csv(path_g7_meta)
g7_result_meta = g7_result_meta.rename(columns={'prediction_lr':'meta_pre_g7'})
g7_path_video = 'Group7_Trailer.csv'
#movie_data =pd.read_csv(path, sep='\t', head = 0)
g7_result_video = pd.read_csv(g7_path_video)
g7_result_video = g7_result_video.drop(columns=['url'])
g7_result_video = g7_result_video.rename(columns={'prediction':'video_pre_g7'})
g7_result = pd.merge(g7_result_meta, g7_result_video, how ='outer',on='imdb_id')
g7_result = g7_result.drop(columns=['imdb_id','prediction'])
g7_result_final = pd.merge(g6_result_final,g7_result,how='outer', on='url')
g7_result_final = g7_result_final.drop_duplicates(subset='url', keep='first', inplace=False)
g7_result_final
```

```python
path_g8 = 'Group8.csv'
g8_result = pd.read_csv(path_g8)
g8_result = g8_result.rename(columns={'prediction':'meta_pre_g8','prediction_video':'video_pre_g8'})
g8_result = g8_result.drop(columns=['imdb_id'])
g8_result_final = pd.merge(g7_result_final,g8_result,how='outer', on='url')
g8_result_final = g8_result_final.drop_duplicates(subset='url', keep='first', inplace=False)
g8_result_final = g8_result_final.fillna(0)#!!!!!!!!!!!nan is repalced by 0, is there any influence?
g8_result_final
```

```python
import numpy as np
#g8_result_final["meta_res_g1"] = g8_result_final["meta_pre_g1"] - g8_result_final["ground_truth"]
#g8_result_final["video_res_g1"] = g8_result_final["video_pre_g1"] - g8_result_final["ground_truth"]
meta_pre = ['meta_pre_g1', 'meta_pre_g2', 'meta_pre_g3', 'meta_pre_g4', 'meta_pre_g5', 'meta_pre_g6', 'meta_pre_g7', 'meta_pre_g8']
video_pre = ['video_pre_g1', 'video_pre_g2', 'video_pre_g3', 'video_pre_g4', 'video_pre_g5', 'video_pre_g6', 'video_pre_g7', 'video_pre_g8']
meta_res_log = ['meta_res_log_g1', 'meta_res_log_g2', 'meta_res_log_g3', 'meta_res_log_g4', 'meta_res_log_g5', 'meta_res_log_g6', 'meta_res_log_g7', 'meta_res_log_g8']
video_res_log = ['video_res_log_g1', 'video_res_log_g2', 'video_log_res_g3', 'video_res_log_g4', 'video_res_log_g5', 'video_res_log_g6', 'video_res_log_g7', 'video_res_log_g8']

for i in range(len(meta_pre)):
    g8_result_final[meta_res_log[i]] = np.log1p(abs(g8_result_final[meta_pre[i]] - g8_result_final["ground_truth"]))
    g8_result_final[video_res_log[i]] = np.log1p(abs(g8_result_final[video_pre[i]] - g8_result_final["ground_truth"]))

g8_result_final
```

```python
import plotly.graph_objects as go
N = 8
c = ['hsl('+str(h)+',50%'+',50%)' for h in np.linspace(0, 360, N)]
d = ['hsl('+str(h)+',30%'+',30%)' for h in np.linspace(0, 360, N)]
#fig = go.Figure(data=[go.Box(
#    y=3.5 * np.sin(np.pi * i/N) + i/N + (1.5 + 0.5 * np.cos(np.pi*i/N)) * np.random.rand(10),
#    marker_color=c[i]
#    ) for i in range(int(N))])

group = ['Group1', 'Group2', 'Group3', 'Group4',
     'Group5', 'Group6', 'Group7', 'Group8']

fig = go.Figure()
for i in range(int(N)):
    fig.add_trace(go.Box(
        x=g8_result_final[meta_res_log[i]],
        y=[group[i]]*len(g8_result_final[meta_res_log[i]]),
        name='Logarithm of meta residuals of Group ''{}'.format(i),
        marker_color=c[i]
    ))
    fig.add_trace(go.Box(
        x=g8_result_final[video_res_log[i]],
        y=[group[i]]*len(g8_result_final[video_res_log[i]]),
        name='Logarithm of video residuals of Group ''{}'.format(i),
        marker_color=d[i]
    ))   


fig.update_layout(
    xaxis=dict(title='The absolute logarithm of residuals', zeroline=False),
    boxmode='group'
)

fig.update_traces(orientation='h') # horizontal box plots
fig.show()
```
